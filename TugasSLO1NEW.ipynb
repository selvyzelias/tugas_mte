```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import math
from scipy import stats as st
```

```python
df=pd.read_csv ("epl-goalScorer(20-21).csv")
df.head()
```

```python
#10 Player terbanyak mencetak gol
df_sorted = df.sort_values('goals', ascending=False)
df_top10goal = df_sorted.head(10)
print(df_top10goal[['player_name', 'goals']])
```

```python
#10 player terlama durasi bermain
df_sorted = df.sort_values('time', ascending=False)
df_top10t= df_sorted.head(10)
print(df_top10t[['player_name', 'time']])
```

```python
#10 player terbanyak assist
df_sorted = df.sort_values('assists', ascending=False)
df_top10assist = df_sorted.head(10)
print(df_top10assist[['player_name', 'assists']])
```

```python
#4 player pencetak gol terbanyak dari kesebelasan manchester city
df_Manchester = df[df['team_title'] == 'Manchester City']
df_Manchester_sorted = df_Manchester.sort_values('goals', ascending=False)
df_top4_Manchester = df_Manchester_sorted.nlargest(4, 'goals')
#print(df_top4_Manchester City)
df_top4_Manchester = df_top4_Manchester.nlargest(4, 'goals')
print(df_top4_Manchester[['player_name','team_title', 'goals']])
```

```python
#kesebelasan mana pencetak gol terbanyak
df_goals_per_team = df.groupby('team_title')['goals'].sum()
df_goals_per_team_sorted = df_goals_per_team.sort_values(ascending=False)
tim_dengan_goal_terbanyak = df_goals_per_team_sorted.index[0]
print(tim_dengan_goal_terbanyak)
```

```python
#Jumlah kartu kuning dan merah yang diterima pembuat assist  terbanyak?
df_assists = df.groupby('player_name')['assists'].sum()
df_assists_sorted = df_assists.sort_values(ascending=False)
top_assister = df_assists_sorted.index[0]
df_top_assist = df[df['player_name'] == top_assister]
total_cards = df_top_assist['red_cards'].sum() + df_top_assist['yellow_cards'].sum()
print(total_cards)
```

```python
#Siapa yang jumlah assist dan golnya paling banyak dan dari kesebelasan mana?
df['total_assists'] = df.groupby('player_name')['assists'].transform('sum')
df['total_goals'] = df.groupby('player_name')['goals'].transform('sum')
df_sorted = df.sort_values(by=['total_assists', 'total_goals'], ascending=False)
df_result = df_sorted[['player_name', 'total_assists', 'total_goals', 'team_title']].head(5)
print(df_result.to_string())
```

```python
#5 kesebelasan yang menyumbangkan gol terbanyak
df_team_goals = df.groupby('team_title')['goals'].sum()
df_team_goals_sorted = df_team_goals.sort_values(ascending=False)
df_result = df_team_goals_sorted.head(5)
print(df_result.to_string())


#Urutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali!
df['total_cards'] = df['red_cards'] + df['yellow_cards']
df_min_cards = df.loc[df['total_cards'] == df['total_cards'].min()]
df_min_cards_sorted = df_min_cards.sort_values(by='goals', ascending=False)
df_result = df_min_cards_sorted[['player_name', 'total_cards', 'total_goals', 'team_title']]
print(df_result.to_string())
```
